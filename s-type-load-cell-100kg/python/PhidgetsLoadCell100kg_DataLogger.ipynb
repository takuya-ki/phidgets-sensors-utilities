```python
import sys
import time 
from Phidget22.Devices.VoltageRatioInput import *
from Phidget22.PhidgetException import *
from Phidget22.Phidget import *
from Phidget22.Net import *

try:
    ch = VoltageRatioInput()
except RuntimeError as e:
    print("Runtime Exception %s" % e.details)
    print("Press Enter to Exit...\n")
    readin = sys.stdin.read(1)
    # exit(1)

def VoltageRatioInputAttached(e):
    try:
        attached = e
        print("\nAttach Event Detected (Information Below)")
        print("===========================================")
        print("Library Version: %s" % attached.getLibraryVersion())
        print("Serial Number: %d" % attached.getDeviceSerialNumber())
        print("Channel: %d" % attached.getChannel())
        print("Channel Class: %s" % attached.getChannelClass())
        print("Channel Name: %s" % attached.getChannelName())
        print("Device ID: %d" % attached.getDeviceID())
        print("Device Version: %d" % attached.getDeviceVersion())
        print("Device Name: %s" % attached.getDeviceName())
        print("Device Class: %d" % attached.getDeviceClass())
        print("\n")

    except PhidgetException as e:
        print("Phidget Exception %i: %s" % (e.code, e.details))
        print("Press Enter to Exit...\n")
        readin = sys.stdin.read(1)
        # exit(1)   
    
def VoltageRatioInputDetached(e):
    detached = e
    try:
        print("\nDetach event on Port %d Channel %d" % (detached.getHubPort(), detached.getChannel()))
    except PhidgetException as e:
        print("Phidget Exception %i: %s" % (e.code, e.details))
        print("Press Enter to Exit...\n")
        readin = sys.stdin.read(1)
        # exit(1)   

def ErrorEvent(e, eCode, description):
    print("Error %i : %s" % (eCode, description))

def VoltageRatioChangeHandler(e, voltageRatio):
    print("VoltageRatio: %f" % voltageRatio)

def SensorChangeHandler(e, sensorValue, sensorUnit):
    print("Sensor Value: %f" % sensorValue)

try:
    ch.setOnAttachHandler(VoltageRatioInputAttached)
    ch.setOnDetachHandler(VoltageRatioInputDetached)
    ch.setOnErrorHandler(ErrorEvent)

    ch.setOnVoltageRatioChangeHandler(VoltageRatioChangeHandler)
    ch.setOnSensorChangeHandler(SensorChangeHandler)

    # Please review the Phidget22 channel matching documentation for details on the device
    # and class architecture of Phidget22, and how channels are matched to device features.

    # Specifies the serial number of the device to attach to.
    # For VINT devices, this is the hub serial number.
    #
    # The default is any device.
    #
    # ch.setDeviceSerialNumber(<YOUR DEVICE SERIAL NUMBER>)
    
    # Have to confirm the serial number using Phidget22 Control Pannel
    ch.setDeviceSerialNumber(469879) # Number of PhidgetBridge 4-Input

    # For VINT devices, this specifies the port the VINT device must be plugged into.
    #
    # The default is any port.
    #
    # ch.setHubPort(0)

    # Specifies that the channel should only match a VINT hub port.
    # The only valid channel id is 0.
    #
    # The default is 0 (false), meaning VINT hub ports will never match
    #
    # ch.setIsHubPortDevice(1)

    # Specifies which channel to attach to.  It is important that the channel of
    # the device is the same class as the channel that is being opened.
    #
    # The default is any channel.
    #
    # ch.setChannel(0)
    ch.setChannel(0) # Which channel do you want to access? In case of PhidgetBridge 4-Input: 0 ~ 3

    # In order to attach to a network Phidget, the program must connect to a Phidget22 Network Server.
    # In a normal environment this can be done automatically by enabling server discovery, which
    # will cause the client to discovery and connect to available servers.
    #
    # To force the channel to only match a network Phidget, set remote to 1.
    #
    # Net.enableServerDiscovery(PhidgetServerType.PHIDGETSERVER_DEVICE);
    # ch.setIsRemote(1)

    print("Waiting for the Phidget VoltageRatioInput Object to be attached...")
    ch.openWaitForAttachment(5000)
except PhidgetException as e:
    print("Phidget Exception %i: %s" % (e.code, e.details))
    print("Press Enter to Exit...\n")
    readin = sys.stdin.read(1)
    # exit(1)

if(ch.getChannelSubclass() == ChannelSubclass.PHIDCHSUBCLASS_VOLTAGERATIOINPUT_BRIDGE):
    ch.setBridgeEnabled(1)

print("Gathering data for 10 seconds...")
time.sleep(10)

try:
    ch.close()
except PhidgetException as e:
    print("Phidget Exception %i: %s" % (e.code, e.details))
    print("Press Enter to Exit...\n")
    readin = sys.stdin.read(1)
    # exit(1) 
print("Closed VoltageRatioInput device")
# exit(0)

```

# ForceDataLogging

```python
import os
import sys
import time
from Phidget22.Devices.VoltageRatioInput import *
from Phidget22.PhidgetException import *
from Phidget22.Phidget import *
from Phidget22.Net import *

# for graph
import matplotlib.pyplot as plt
from pylab import *
%matplotlib inline

# for handling data
import pandas as pd
import numpy as np
from datetime import datetime as dt

# for file operation
import csv

serialNum = 469879 # PhidgetBridge 4-Input
loadCellNo_ch = ['1003','0861','0588']
loadCell_RatedOut = {'1003':3.0020,'0861':2.9981,'0588':2.9987}
capacity = 100 # Load Cell Capacity [kg]
Data = {} # data
coef = {} # Proportionality coefficient: Capacity/Rated Output
offset = {} # Offset

for LCno in loadCellNo_ch:
    Data['volratio'+LCno] = np.empty(0,dtype=int)
    Data['force'+LCno] = np.empty(0,dtype=int)
    Data['datetime'+LCno] = []
    Data['time'+LCno] = [0]
    coef[LCno] = capacity/loadCell_RatedOut[LCno]
    offset[LCno] = 0

def VoltageRatioInputAttached(e):
    try:
        attached = e
        print("\nAttach Event Detected (Information Below)")
        print("===========================================")
        print("Library Version: %s" % attached.getLibraryVersion())
        print("Serial Number: %d" % attached.getDeviceSerialNumber())
        print("Channel: %d" % attached.getChannel())
        print("Channel Class: %s" % attached.getChannelClass())
        print("Channel Name: %s" % attached.getChannelName())
        print("Device ID: %d" % attached.getDeviceID())
        print("Device Version: %d" % attached.getDeviceVersion())
        print("Device Name: %s" % attached.getDeviceName())
        print("Device Class: %d" % attached.getDeviceClass())
        print("\n")

    except PhidgetException as e:
        print("Phidget Exception %i: %s" % (e.code, e.details))
        print("Press Enter to Exit...\n")
        readin = sys.stdin.read(1)
        # exit(1)   
    
def VoltageRatioInputDetached(e):
    detached = e
    try:
        print("\nDetach event on Port %d Channel %d" % (detached.getHubPort(), detached.getChannel()))
    except PhidgetException as e:
        print("Phidget Exception %i: %s" % (e.code, e.details))
        print("Press Enter to Exit...\n")
        readin = sys.stdin.read(1)
        # exit(1)   

def ErrorEvent(e, eCode, description):
    print("Error %i : %s" % (eCode, description))

def VoltageRatioChangeHandler_1003(e, voltageRatio):
    global Data
    Data['datetime1003'].append(dt.now())
    Data['volratio1003'] = np.append(Data['volratio1003'],np.array(voltageRatio))
    #print("VoltageRatio(1003): %f" % voltageRatio)
    #print("MeasuredForce(1003): %f" % (coef['1003'] * voltageRatio + offset['1003']))

def VoltageRatioChangeHandler_0861(e, voltageRatio):
    global Data # For data storage
    Data['datetime0861'].append(dt.now())
    Data['volratio0861'] = np.append(Data['volratio0861'],np.array(voltageRatio))
    #print("VoltageRatio(0861): %f" % voltageRatio)
    #print("MeasuredForce(0861): %f" % (coef['0861'] * voltageRatio + offset['0861']))

def VoltageRatioChangeHandler_0588(e, voltageRatio):
    global Data # For data storage
    Data['datetime0588'].append(dt.now())
    Data['volratio0588'] = np.append(Data['volratio0588'],np.array(voltageRatio))
    #print("VoltageRatio(0588): %f" % voltageRatio)
    #print("MeasuredForce(0588): %f" % (coef['0588'] * voltageRatio + offset['0588']))

def SensorChangeHandler(e, sensorValue, sensorUnit):
    print("Sensor Value: %f" % sensorValue)

def SetChannel(hubSerialNum):
    ch = [0]*len(loadCellNo_ch)
    for chNum in range(len(loadCellNo_ch)):
        # Create Instance of VoltageRatioInput
        try:
            ch[chNum] = VoltageRatioInput()
        except RuntimeError as e:
            print("Runtime Exception %s" % e.details)
            print("Press Enter to Exit...\n")
            readin = sys.stdin.read(1)

        # Setting about connection
        try:
            ch[chNum].setOnAttachHandler(VoltageRatioInputAttached)
            ch[chNum].setOnDetachHandler(VoltageRatioInputDetached)
            ch[chNum].setOnErrorHandler(ErrorEvent)

            if loadCellNo_ch[chNum] == '1003':
                ch[chNum].setOnVoltageRatioChangeHandler(VoltageRatioChangeHandler_1003)
            elif loadCellNo_ch[chNum] == '0861':
                ch[chNum].setOnVoltageRatioChangeHandler(VoltageRatioChangeHandler_0861)
            elif loadCellNo_ch[chNum] == '0588':
                ch[chNum].setOnVoltageRatioChangeHandler(VoltageRatioChangeHandler_0588)

            ch[chNum].setOnSensorChangeHandler(SensorChangeHandler)

            ch[chNum].setDeviceSerialNumber(hubSerialNum) # Number of PhidgetBridge 4-Input
            ch[chNum].setChannel(chNum) # Which channel do you want to access? In case of PhidgetBridge 4-Input: 0 ~ 3

            # In order to attach to a network Phidget, the program must connect to a Phidget22 Network Server.
            # In a normal environment this can be done automatically by enabling server discovery, which
            # will cause the client to discovery and connect to available servers.
            #
            # To force the channel to only match a network Phidget, set remote to 1.
            #
            # Net.enableServerDiscovery(PhidgetServerType.PHIDGETSERVER_DEVICE);
            # ch.setIsRemote(1)

            print("Waiting for the Phidget VoltageRatioInput Object to be attached...")
            ch[chNum].openWaitForAttachment(5000)
        except PhidgetException as e:
            print("Phidget Exception %i: %s" % (e.code, e.details))
            print("Press Enter to Exit...\n")
            readin = sys.stdin.read(1)
            # exit(1)

        # Gain setting and Resolution
        ch[chNum].setBridgeGain(1)
        # Discription about Data_Interval/Change_Trigger
        # https://www.phidgets.com/docs/Data_Interval/Change_Trigger
        ch[chNum].setVoltageRatioChangeTrigger(0.0)
        ch[chNum].setDataInterval(8) # Default: 256 ms(8ms ~ 1000ms)(8ms, 122 samples a second)
        time.sleep(1)
        
    return ch

def DatetimeToTime(vrd,lenvrd):
    sum_time = 0
    dtime = [0]
    for i in range(1,lenvrd):
        sum_time += (vrd[i]-vrd[i-1]).total_seconds()
        dtime.append(sum_time)
    return dtime

def Calibrate(ch):
    print("Calibrating...\n")
    time.sleep(5) # Wait for X seconds
    ReleaseChannel(ch)
    
    # Calculate offset
    global offset, coef, Data
    for LCno in loadCellNo_ch:
        lenvrd = len(Data['datetime'+LCno])
        offset[LCno] = 0 - coef[LCno] * np.average(Data['volratio'+LCno][lenvrd/5:lenvrd])
        Data['time'+LCno] = DatetimeToTime(Data['datetime'+LCno],lenvrd)
    
    print("\nFiniched Calibration!\n")
    
def StartLogging(ch):
    for i in range(len(loadCellNo_ch)):
        if(ch[i].getChannelSubclass() == ChannelSubclass.PHIDCHSUBCLASS_VOLTAGERATIOINPUT_BRIDGE):
            ch[i].setBridgeEnabled(1)

def RestartLogging():
    print("***** Restart Data Logging *****\n")
    
    # Initialize
    for LCno in loadCellNo_ch:
        Data['datetime'+LCno] = []
        Data['time'+LCno] = [0]
        Data['volratio'+LCno] = np.empty(0,dtype=int)
    ch = SetChannel(serialNum)
    StartLogging(ch)
    print(dt.now())
    return ch

def PlotSaveResults():
    # Calculate force value
    for LCno in loadCellNo_ch:
        fd = Data['force'+LCno] # Copy of object(mutable object)
        vd = Data['volratio'+LCno]*1000
        fd = 9.8 * coef[LCno] * vd + offset[LCno] # [N]
        
        Data['time'+LCno] = DatetimeToTime(Data['datetime'+LCno],len(Data['datetime'+LCno]))
        
        # Plot
        resultfig = plt.figure(figsize=(16,4))
        resultax = resultfig.add_subplot(1,1,1)
        resultax.set_ylabel('Measured Force [N]',fontsize=14)
        resultax.set_xlabel('Time [sec]',fontsize=14)
        resultax.plot(Data['time'+LCno],fd,linewidth=2,color='red',alpha=0.6)
        plt.title('Results',size=14)
        
        # Save
        with open('results.csv', 'w') as f:
            writer = csv.writer(f, lineterminator='\n') # Specify a line feed code (\n)
            writer.writerow(['Time[sec]','MeasuredForce[N]']) # Write a header
            writer.writerows(np.c_[Data['time'+LCno],fd]) # Write a data
    print("\nSaved results")
            
    
def ReleaseChannel(ch):
    # Release the channel on the Phidget device
    for i in range(len(ch)):
        try:
            ch[i].close()
        except PhidgetException as e:
            print("Phidget Exception %i: %s" % (e.code, e.details))
            print("Press Enter to Exit...\n")
            readin = sys.stdin.read(1)
            # exit(1) 
    print("Closed VoltageRatioInput device")
    # exit(0)

if __name__ == '__main__':
    
    # Comprehensive is too early
    ch = SetChannel(serialNum)
    # Here, starting logging data
    StartLogging(ch)
    # Calibrate sensor
    Calibrate(ch)
    # Restart logging data
    ch = RestartLogging()
    
    while 1:
        key = input()
        # End with zDisplayed if there is another key input
        if key == 'z':
            break
        elif key:
            print(key)
    
    # Plot force
    ReleaseChannel(ch)
    PlotSaveResults()
```
